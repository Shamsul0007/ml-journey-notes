Which model predictes better to identify the Delaney Solubility?

```python
import pandas as pd
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
```

```python
y = df['logS']
y
```

```python
X= df.drop('logS', axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_lr_train_ped= lr.predict(X_train)
y_lr_test_ped= lr.predict(X_test)
```

```python
y_lr_train_ped
```

```python
y_lr_test_ped
```

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse= mean_squared_error(y_train, y_lr_train_ped)
lr_train_r2= r2_score(y_train, y_lr_train_ped)

lr_test_mse= mean_squared_error(y_test, y_lr_test_ped)
lr_test_r2= r2_score(y_test, y_lr_test_ped)


```

```python
print ('LR_MSE (Train): ', lr_train_mse)
print ('LR_MSE (Test): ', lr_test_mse)
print ('LR_R2 (Train): ', lr_train_r2)
print ('LR_R2 (Test): ', lr_test_r2)
```

```python
lr_results= pd.DataFrame(['Linear regression',lr_train_mse, lr_train_r2,lr_test_mse, lr_test_r2]).transpose()
lr_results.columns= ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

```python
from sklearn.ensemble import RandomForestRegressor
rf= RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

```python
y_rf_train_ped= rf.predict(X_train)
y_rf_test_ped= rf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse= mean_squared_error(y_train, y_rf_train_ped)
rf_train_r2= r2_score(y_train, y_rf_train_ped)

rf_test_mse= mean_squared_error(y_test, y_rf_test_ped)
rf_test_r2= r2_score(y_test, y_rf_test_ped)


```

```python
rf_results= pd.DataFrame(['Random Forest',rf_train_mse, rf_train_r2,rf_test_mse, rf_test_r2]).transpose()
rf_results.columns= ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
df_models= pd.concat([lr_results, rf_results], axis=0).reset_index()
df_models
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figszie=(5,5)
plt.scatter(x=y_train, y=y_lr_train_ped, c="#7CAE00", alpha=0.4)

z=np.polyfit(y_train, y_lr_train_ped, 1)
p= np.poly1d(z)

plt.plot(y_train, p(y_train), '#F8766D')
plt.ylabel('Predict LogS')
plt.xlabel('Experimental LogS')
```

Conclusion: Linear Regression performs better on the test set in terms of both lower MSE and higher R², making it the better choice for this task based on current results.
