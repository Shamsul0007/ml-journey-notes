```python
import pandas as pd
df= pd.read_csv('/content/vgsales.csv')
df
```

```python
df.describe()
```

```python
import numpy as np
```

```python
df['Year']=df['Year'].replace('N/A',np.nan)
```

```python
median_year=df['Year'].median()
```

```python
df['Year'].fillna(median_year,inplace=True)
```

```python
df.describe()
```

```python
y= df['Global_Sales']
y
```

```python
X= df.drop(columns=['Rank','Name','Global_Sales'],axis=1)
X
```

```python
X_encoded = pd.get_dummies(X, columns=['Platform', 'Genre', 'Publisher'], drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_encoded,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_lr_train_ped= lr.predict(X_train)
y_lr_test_ped= lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse= mean_squared_error(y_train, y_lr_train_ped)
lr_train_r2= r2_score(y_train, y_lr_train_ped)

lr_test_mse= mean_squared_error(y_test, y_lr_test_ped)
lr_test_r2= r2_score(y_test, y_lr_test_ped)
```

```python
lr_results= pd.DataFrame(['Linear regression',lr_train_mse, lr_train_r2,lr_test_mse, lr_test_r2]).transpose()
lr_results.columns= ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
lr_results
```
